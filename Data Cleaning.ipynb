```python
#libraries
import pandas as pd
import re
import string
import unicodedata

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('punkt')
from nltk.stem.porter import PorterStemmer

import random
import numpy as np
from tqdm import tqdm
```

## Data Cleaning

```python
df = pd.read_csv(r"data/fake_job_postings.csv")
df.head()
```

```python
#Fill missing values with blank
df.interpolate(inplace=True)

#Check total number of null in each attributes
df.isnull().sum()

#Fill NaN values with blank space
df.fillna(' ', inplace=True)

df.head()
```

```python
# create the copy of dataframe
df2 = df.copy()

#remove columns thats irrelevent 

df2 = df2.drop(['job_id', 'salary_range','telecommuting', 'has_company_logo', 
                'has_questions', 'department', 'function'], axis=1)

#merge the columns into one column for processing aside from fraudulent

df2['text'] = df2['title']+'. '+ df2['location']+'. '+ df2['company_profile']+'. '+ df2['description']+'. '+ df2['requirements']+'. '+ df2['benefits'] + '. '+ df2['employment_type'] + '. '+ df2['required_experience']+ '. '+ df2['required_education']+ '. '+ df2['industry']

df2.head()
```

```python
#delete the other columns

df2 = df2.drop(['title', 'location','company_profile','description','requirements',
                'benefits','employment_type','required_experience','required_education','industry'], axis=1)
```

```python
# convert to lowercase
df2['clean_text'] = df2['text'].astype(str).str.lower()
```

```python
df2
```

```python
# expand contradictions
def expand(text):
    text = re.sub(r"won\'t", " will not", text)
    text = re.sub(r"can\'t", " can not", text)
    text = re.sub(r"n\'t", " not", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"\'s", " is", text)
    text = re.sub(r"\'d", " would", text)
    text = re.sub(r"\'ll", " will", text)
    text = re.sub(r"\'t", " not", text)
    text = re.sub(r"\'ve", " have", text)
    text = re.sub(r"\'m", " am", text)
    return text

df2['clean_text'] = df2['clean_text'].apply(expand)
```

```python
def cleaning(text):
    text = re.sub(r'[^a-zA-z0-9\s]', ' ', text) #remove special characters
    text = re.sub('[%s]'% re.escape(string.punctuation),' ', text) #remove punctuation
    text = re.sub(r'\d+', ' ', text) #remove numbers
    text = re.sub(r'\s+',' ',text) # replace multiple spaces with only one space
    return text

df2['clean_text'] = df2['clean_text'].apply(cleaning)
```

```python
# remove accented characters

def accented_chars(text):
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    return text

df2['clean_text'] = df2['clean_text'].apply(accented_chars)
```

```python
stemmer = PorterStemmer()
 
def stemming(text):
    stems = [stemmer.stem(word) for word in text]
    return stems

df2['clean_text'] = df2['clean_text'].apply(stemming)

#df2.iloc[0]['clean_text']
```

```python
df2 = df2.drop(['text'], axis=1)
df2.to_csv("dataset_cleaned.csv",index=False)
df2.head()
```
