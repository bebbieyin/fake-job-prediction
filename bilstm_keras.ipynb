```python
import pandas as pd

import tensorflow as tf

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.callbacks import EarlyStopping, ModelCheckpoint

import sklearn
from sklearn.metrics import confusion_matrix,classification_report,confusion_matrix,accuracy_score

import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
df_clean = pd.read_csv('data/dataset_cleaned.csv')
df_clean.head()
```

```python
neg, pos = np.bincount(df_clean['fraudulent'])
total = neg + pos
print('Examples:\n    Total: {}\n    Positive: {} ({:.2f}% of total)\n'.format(
    total, pos, 100 * pos / total))
```

## Transform words into features

Models cannot directly input words, we need to transform it to numerical features for it to process. 

```python
# HYPERPARAMETERS

BATCH_SIZE =64
LR_RATE = 1e-4
EPOCH = 50

vocab_size = 100000
embedding_dim = 50
max_length = 300
```

```python
#create tokenizer
tokenizer = Tokenizer(num_words=vocab_size)

# create a dictionary of all the words in the dataset
tokenizer.fit_on_texts(df_clean['clean_text'].values)
word_index = tokenizer.word_index
word_index
```

```python
# transforms the sentences into features
X = tokenizer.texts_to_sequences(df_clean['clean_text'].values)     

# pad the sentences so they are equal length
X = pad_sequences(X, maxlen=max_length)

# assign Y value
Y=df_clean['fraudulent']                                                                   

print("Original sentence:")
print(df_clean.iloc[0]['clean_text'])
print()
print("Transformed to numerical features:")
print(X[0])
```

## Class weights

One way to handle imbalanced data is to balance the class weights for the two classes during training. We can input the weights of each class to the model so that it will give more importance to the minority class. Class weight for each class can be calculated using the below formula: <br>

total samples / (no.of classes * no. of samples in that class)

```python
# Scaling by total/2 helps keep the loss to a similar magnitude.
# The sum of the weights of all examples stays the same.
weight_for_0 = total/(2 * neg)
weight_for_1 = total/(2 * pos)

class_weights = {0: weight_for_0, 1: weight_for_1}

print('Weight for class 0: {:.2f}'.format(weight_for_0))
print('Weight for class 1: {:.2f}'.format(weight_for_1))
```

```python
from sklearn.model_selection import train_test_split

# 80% training data, 10% validation data, 10% for test
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.10,random_state=41)

print("Class distribution in testing set",np.bincount(y_test))
```

```python
METRICS = [
      tf.keras.metrics.TruePositives(name='tp'),
      tf.keras.metrics.FalsePositives(name='fp'),
      tf.keras.metrics.TrueNegatives(name='tn'),
      tf.keras.metrics.FalseNegatives(name='fn'), 
      tf.keras.metrics.BinaryAccuracy(name='accuracy'),
      tf.keras.metrics.Precision(name='precision'),
      tf.keras.metrics.Recall(name='recall'),
      tf.keras.metrics.AUC(name='auc'),
      tf.keras.metrics.AUC(name='prc', curve='PR'), # precision-recall curve
]
```

Note: Using class_weights changes the range of the loss. This may affect the stability of the training depending on the optimizer. Optimizers whose step size is dependent on the magnitude of the gradient, like tf.keras.optimizers.SGD, may fail. The optimizer used here, tf.keras.optimizers.Adam, is unaffected by the scaling change.

**Model Architecture**<br /><br />

***Embedding Layer***<br /><br />
The first layer is an embedding layer that is widely used in NLP tasks. From the above section we mentioned that the model can only process numbers not strings, thus we tokenize each word and assign an unique interger to it, each word is regarded as an unique feature. For example, the number 1 will be assigned to the word "hire", the number 2 will be assigned to the word "experience", and so on for each word in the dataset. <br />

But the limitation of this type of encoding is there is no relationship being from between the words. Similar words are not closer together and vice versa. That makes it harder for the model to interpret it as well as neural networks learn the a single weights for each feature and combine them those weights in order for it to learn. <br />

Therefore, we have to transform these features(words) into word embeddings. Word embeddings are typically N-dimensional floating point values used to represent each word. Words with similar meaning will have a similar representation. In this example, we chose the embedding dimension as 50. 

These embeddings are trainable, that means that the model is able learn how to better represent each word as it improves. <br /><br />


***Bidirectional LSTM Layer***<br /><br />
Two LSTM with 32 units is used, one in both direction. Resulting in a total of 64 units in the final model.<br /><br />

***Dropout Layer***<br /><br />
Dropout is used to prevent overfitting. It randomly ignores some of the neurons in the network and prevent it for contributing to the next layer.<br /><br />

***Dense Layer***<br /><br />
Dense layer also known as fully connected layer takes all outputs from neuron of the previous layers. The dropout rate in the last layer is set to 0.5, therefore only half of the neurons of the previous layer actually contributes to this. It is used as final classifier layer for our model, with sigmoid as the activation function as we are working with binary classification task.

```python
model = tf.keras.Sequential([
        tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(learning_rate=LR_RATE),
              weighted_metrics =METRICS)
model.summary()
```

```python
#callbacks = EarlyStopping(monitor='val_prc', patience=10,restore_best_weights=True)

checkpoint_path = "checkpoints/weights_{epoch:03d}.h5"
checkpoint_dir = os.path.dirname(checkpoint_path)

checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath=checkpoint_path,
    save_weights_only=True,
    monitor='val_prc',
    mode='max',
    )
```

```python

history = model.fit(X_train, y_train, epochs=EPOCH,batch_size=BATCH_SIZE, validation_split=0.1, shuffle=True,verbose=1,
                   callbacks=[checkpoint_callback],class_weight=class_weights)
```

## Training Visualization

```python
import json

hist_path = "checkpoints/history.json"
history_dict = history.history
json.dump(history_dict, open(hist_path, 'w'))
```

```python
history_dict = json.load(open(hist_path, 'r'))
best_prc = max(history_dict['val_prc'])
index = history_dict['val_prc'].index(best_prc)+1
print("Best validation PRC "+str(max(history_dict['val_prc'])))
print("Best validation Precision-Recall is at epoch "+str(index))
```

```python
plt.figure(figsize=(10, 10), dpi=80)

def plot_metrics(history):

    metrics = ['loss', 'prc', 'precision', 'recall']
    for n, metric in enumerate(metrics):
        num_epochs = len(history_dict[metric])
        epochs=list(np.arange(0,num_epochs))

        name = metric.replace("_"," ").capitalize()
        plt.subplot(2,2,n+1)
        plt.plot(epochs, history[metric], label='Train')
        plt.plot(epochs, history['val_'+metric],
                  linestyle="--", label='Val')
        plt.xlabel('Epoch')
        plt.ylabel(name)
        if metric == 'loss':
            plt.ylim([0, plt.ylim()[1]])
        elif metric == 'auc':
            plt.ylim([0.8,1])
        else:
            plt.ylim([0,1])
            

    plt.legend()
    
plot_metrics(history_dict)
```

## Evaluation

```python
plt.figure(figsize=(10, 10), dpi=80)

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
def plot_cm(labels, predictions, p=0.5):
    cm = confusion_matrix(labels, predictions > p)
    plt.figure(figsize=(5,5))
    sns.heatmap(cm, annot=True, fmt="d")
    plt.title('Confusion matrix @{:.2f}'.format(p))
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')

    print('Real Jobs Detected (True Negatives): ', cm[0][0])
    print('Real Jobs Incorrectly Detected (False Positives): ', cm[0][1])
    print('Fraudulent Jobs Missed (False Negatives): ', cm[1][0])
    print('Fraudulent Jobs Detected (True Positives): ', cm[1][1])
    print('Total Fraudulent Jobs in Test Set: ', np.sum(cm[1]))

def plot_prc(name, labels, predictions, **kwargs):
    precision, recall, _ = sklearn.metrics.precision_recall_curve(labels, predictions)

    plt.plot(precision, recall, label=name, linewidth=2, **kwargs)
    plt.xlabel('Precision')
    plt.ylabel('Recall')
    plt.grid(True)
    ax = plt.gca()
    ax.set_aspect('equal')
```

```python
test_predictions_weighted = model.predict(X_test, batch_size=BATCH_SIZE)
```

```python


weighted_results = model.evaluate(X_test, y_test,batch_size=BATCH_SIZE, verbose=0)

for name, value in zip(model.metrics_names, weighted_results):
    
    if name=="precision":
        precision=value
    if name=="recall":
        recall=value
        
    print(name, ': ', value)
  


f1_score = 2 * (precision * recall) / (precision + recall)

print("F1 score : "+str(f1_score))
print()

plot_cm(y_test, test_predictions_weighted)
```

```python
# PRC curve of testing
plot_prc("Test", y_test, test_predictions_weighted, color=colors[1], linestyle='--')


plt.legend(loc='lower right');
```
