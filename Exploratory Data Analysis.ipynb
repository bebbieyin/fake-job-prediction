# Importing Data

**<ins>Dataset Description**</ins>

Originally, this dataset is from Employment Scam Aegean Dataset (EMSCAD) where its collected by University of the Aegean, Laboratory of Information & Communication Systems Security. It is made available on Kaggle. It has 18k instances with 18 attributes. The target variable is fraudulent where 0 means real and 1 means fake.

There are 17 categorical , and 1 numerical data.


```python
df = pd.read_csv(r'data/fake_job_postings.csv')
df.head()
```

```python
# data size
print("Dataset size\n")
print(str(df.shape)+'\n')

# type of attributes
print("Types of attributes\n")
print(df.info())
```

```python
#Fill missing values with blank
df.interpolate(inplace=True)
#Fill NaN values with blank space
df.fillna(' ', inplace=True)

#Check total number of null in each attributes
df.isnull().sum()
```

# Data Visualization

### Number of Real vs. Fake Jobs
The bar chart and summary table show that only 5% of the data are fake jobs compared to real jobs. 

```python
# number of class distributions

df['fraudulent'].value_counts().plot(kind='bar',title='No. of Real vs Fake Jobs')
df.groupby('fraudulent').count()['title'].reset_index().sort_values(by='title',ascending=False)
```

# Job Postings in Each Countries

```python
def split(location):
    l = location.split(',')
    return l[0]

df['country'] = df.location.apply(split)

country = dict(df.country.value_counts()[:11])
del country[' ']
plt.figure(figsize =(6, 4))
plt.title('Job Posting by Countries', size=15)
colors = ['#C0C0C0'] * 10
colors[0] = '#029386'
plt.bar(country.keys(), country.values(),color=colors)
plt.ylabel('Number of Jobs')
plt.xlabel('Countries')
```

## Entry requirements for fake and real jobs

From the visualisation, most of the fake jobs only required entry-level experience and rarely required an internship or executive experience to get the job. While for a real job, a mid-senior level is required for most of the jobs. 

```python
#Visualize the required experiences in the fake jobs
experience1 = dict(df[df.fraudulent==1].required_experience.value_counts())
del experience1[' ']
del experience1['Not Applicable']

plt.figure(figsize=(18,4))

plt.subplot(1,2,1)

plt.bar(experience1.keys(), experience1.values(),color = sns.color_palette('Reds_r', 6))
plt.title('Required Experiences in Fake Job', size=15)
plt.xlabel('Experiences')
plt.ylabel('No. of jobs')

#Visualize the required experiences in the real jobs
experience2 = dict(df[df.fraudulent==0].required_experience.value_counts())
del experience2[' ']
del experience2['Not Applicable']

plt.subplot(1,2,2)

plt.bar(experience2.keys(), experience2.values(),color = sns.color_palette('Greens_r', 6))
plt.title('Required Experiences in Real Job', size=15)
plt.xlabel('Experiences')
plt.ylabel('No. of jobs')

plt.show()
```

## Top industries of fake and real jobs</ins>

To visualize which industries have the highest number of fake and real jobs, pie charts are plotted to find the top 5 industries for both. <br>

According to the US Energy and Employment Report 2019, nearly 1.5 million directly employed by the oil and energy industry due to a raise in natural gas production. Those gains have translated into more oil and energy industry jobs. Thus, in the first pie chart, oil & energy shows the highest number of fake job recruitment as it has the highest median salary of any industry in the S&P 500. <br>

The second pie chart shows that information technology & services have the highest number of real job postings as in this industry, it required strong technical skills to get the job. Thus, it has a lesser chance for job seekers that lack tech skills to be scammed.

```python
industry1 = df[df.fraudulent==1].industry.value_counts().head(6)
del industry1[' ']

labels1 = ["Oil & Energy","Accounting","Hospital & Health Care" ,"Marketing & Advertising","Financial Services "]
explode1 = [0.1,0,0,0,0]
colors1 = sns.color_palette('Reds_r', 5)

plt.figure(figsize=(18,4))
plt.subplot(1,2,1)

plt.pie(industry1,labels=labels1,explode=explode1,colors=colors1)
plt.title('Top 5 Industry for Fake Jobs', size=15)

plt.show()

#Pie plot for real jobs
industry2 = df[df.fraudulent==0].industry.value_counts().head(6)
del industry2[' ']

labels2 = ["Information Technology & Services", "Computer Software", "Internet", "Education Management", "Marketing & Advertising"]
explode2 = [0.1,0,0,0,0]
colors2 = sns.color_palette('Greens_r', 5)

plt.figure(figsize=(18,4))
plt.subplot(1,2,2)

plt.pie(industry2,labels=labels2,explode=explode2,colors=colors2)
plt.title('Top 5 Industry for Real Jobs', size=15)

plt.show()
```

## Common words used in real and fake job advertisements

```python
# import the cleaned dataset

df_clean = pd.read_csv(r'data/dataset_cleaned.csv')
df_clean.head()
```

```python
# split sentences into words

tokenizer = Tokenizer(num_words=500000)
tokenizer.fit_on_texts(df_clean['clean_text'].values)
word_index = tokenizer.word_index
```

```python
#Separate fake and actual jobs
fakejobs_text = df_clean[df_clean.fraudulent==1].clean_text
realjobs_text = df_clean[df_clean.fraudulent==0].clean_text

fakejobs_text = ' '.join(str(v) for v in fakejobs_text)
realjobs_text = ' '.join(str(v) for v in realjobs_text)
```

```python
# filter stopwords and noises
STOPWORDS = spacy.lang.en.stop_words.STOP_WORDS
STOPWORDS.update(["time","project","year","required","requirement","requirements","years","customer", "service","experience","b","c","d","e","f"])
```

```python
# Fake jobs word cloud
plt.figure(figsize = (16,14))
wc1 = WordCloud(background_color="white", min_font_size = 10,  max_words = 4000 , width = 1500 , height = 700 , stopwords = STOPWORDS).generate(fakejobs_text)
plt.imshow(wc1,interpolation = 'bilinear')
plt.title("Fake Job Word Cloud")
plt.axis("off")
plt.show()
```

```python
#Actual jobs wordcloud
plt.figure(figsize = (16,14))
wc2 = WordCloud(background_color="white", min_font_size = 10,  max_words = 4000 , width = 1500 , height = 700 , stopwords = STOPWORDS).generate(realjobs_text)
plt.imshow(wc2,interpolation = 'bilinear')
plt.title("Actual Job Word Cloud")
plt.axis("off")
plt.show()
```
